```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
species=["setosa", "versicolor", "virginica"]
size= [5,10,15] #size value should be same as string value
ax.pie(size, labels=species, autopct= "%1.1f%%") #piechart can be drawn only for numerics so put that first
plt.show()           
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
species=["setosa", "versicolor", "virginica"]
size= [5,10,15]
ax.bar(species,size) 
plt.xlabel("species")
plt.ylabel("size of the species") 
plt.show()           
```

```python
fig3 = plt.figure
ax = plt.axes(projection="3d")
z = 15*np.random.random(50)
x = np.random.random(50) #np.sin(z)
y = np.random.random(50) #np.coz(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
